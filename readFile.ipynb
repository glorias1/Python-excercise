```python
import os
f = open('news.txt', 'r', encoding="utf8")
content = f.readlines()
print(content)
```

```python
import re
text = ''
for i in content:
    line = i.replace('\n', '') # remove EOL symbol
    line = line.replace(r'“', '') # remove unnecessary punctuation
    line = line.replace(r'\’', '') # remove unnecessary punctuation
    line = line.replace(r'”', '')   # remove unnecessary punctuation
    line = re.sub(r'[^\w\s]','',line) # remove punctuations
    if line != '':
        text = text + ' ' + line

print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordcloud = WordCloud().generate(text)
# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
def wordCound(text):
    return len(text.split(' '))

print(wordCound(text))
```
